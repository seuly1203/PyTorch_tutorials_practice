# Transforms
- 학습에 적합한 데이터로 만드는데 사용
- 학습을 위해 features들은 normalized된 tensor로, labels은 one-hot encoded tensors로 준비해야함 (FashionMNIST 데아터 기준: features - image, lables - integers)
- 이와 같은 변환을 위해 ToTensor, Lambda 사용

```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

ds = datasets.FashionMNIST(
    root = 'data',
    train = True,
    download = True,
    transform = ToTensor(),
    target_transform = Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)
```

## ToTensor()

- converts a PIL image or NumPy ndarray into a FloatTensor
- and scales the image's pixel intensity values in the range[0., 1.]

## Lambda Transforms

- apply any user-defined lambda function

```python
# using **lambda** we define a function to turn the integer into a one-hot encoded tensor
target_transform = Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1)

# creates a zero tensor of size 10 (the number of labels in the dataset)
Lambda(lambda y: torch.zeros(10, dtype=torch.float))

# calls scatter_ which assigns a value=1 on the index as given by the label y
.scatter_(dim=0, index=torch.tensor(y), value=1)
```
