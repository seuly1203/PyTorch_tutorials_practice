# Save and Load the Model

```python
import torch
import torchvision.models as models
```

# Saving and Loading Model Weights

```python
# PyTorch models store the learned parameters in an internal state dictionary,
# called [state_dict]. These can be persisted via the [torch.save] method
model = models.vgg16(pretrained = True)
torch.save(model.state_dict(), 'model_weights.pth')
```

```python
# to load model weights, creation of an instance of the same model is needed first
# then can load the parameters using [load_state_dict()] method

model = models.vgg16() # pretrained = True 사용하지 않으면 default weight 로드 x
model.load_state_dict(torch.load('model_weights.pth'))
model.eval()
# make sure to call [model.eval()] before inferencing to set the dropout and batch
# normalization layers to evaluation mode -> 안그러면 일치하지 않는 inference results 나옴
```

# Saving and Loading Models with Shapes

- 모델 가중치를 로딩할 때 모델 클래스를 먼저 instantiate해야함: 클래스가 신경망 구조를 정의하기 때문에
- 따라서, 이 클래스의 구조를 모델과 함께 저장하면 편리 in which case we can pass model (and not model.state_dict()) to the saving function: torch.save()
